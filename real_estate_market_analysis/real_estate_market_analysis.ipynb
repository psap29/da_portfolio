### Предобработка данных

#### Заполнение пропусков в данных

При первичном анализе данных выяснили, что ряд колонок содержат не все данные, однако стратегия работы с заполнением пропусков будет разная. **Начнем со столбца `is_apartment`:** все пропуски заполним значением False в силу того, что закономерностей между апартаментами/квартирами по столбцам замечено не было. Более того, из присутствующих объектов только около 2% представляют собой апартаменты, поэтому на проводимом анализе полученные результаты не скажутся (а вот если бы мы заполнили пропуски на основе цены или других столбцов, то, вполне вероятно, что да). В качестве гипотезы будем учитывать то, что владельцы недвижимости попросту оставляют незаполненным данное поле, так как их тип недвижимости не относится к апартаментам. На реальном проекте можно было бы уточнить является ли данный параметр обязательным при заполнении объявления и скорректировать гипотезу.

```python
data['is_apartment'] = data['is_apartment'].fillna('False')
```

**Аналогичную гипотезу представляет собой столбец `balcony`:** вероятно, что пропуски там также равны отсутствию балкона, поэтому заполним пропущенные значения нулями.

```python
data['balcony'] = data['balcony'].fillna(0)
```

**Данные столбца `ceiling_height`** следует сначала проанализировать, так как в них присутствуют выбросы, исходя из анализа гистограммы, принять решение по работе с ними, а уже потом выбрать стратегию заполнения пропусков в данных. Построит box-plot для определения аномальных значений.

```python
data.boxplot('ceiling_height')
```

На графике видно, что есть выбросы. Данные выбросы не случайны и связаны с форматированием, так как часть данных введена неверно (например, 14, что, скорее всего, равно 1.4). Исходя из информации в открытых источников, максимальная высота потолков в СПб достигает 5-6 метров (старый жилой фонд), в новом же - 3, рассчитаем статистику для подтверждения гипотезы.

```python
data['ceiling_height'].describe()
```

Видим, что 75-процентиль равен 2.8, а это значит, что 25% данных составляют собой значения больше 2.8 метров высоты потолков, что подтверждает гипотезу. Для форматирования всех данных напишем цикл, который будет перебирать значения > 10 метров и приводить их к нужному форматированию (высоту в более чем 10 метров представить сложно, а минимальное значение потолка как раз равно 1 метру).

```python
for i in data['ceiling_height']:
     if i >= 10:
         data['ceiling_height'] = data['ceiling_height'].replace({i: i/10})
        
# построим гистограмму и посмотрим на новое распределение
data.hist('ceiling_height')
```

Видим, что теперь выбросов нет и данные распределены достаточно равномерно. Заполним пропуски медианным значением, округлив в большую сторону и проверим, изменилось ли распределение.

```python
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median())
data.hist('ceiling_height', bins=30)
```

Так, исходя из полученной гистограммы, пропуски были заполнены верно, так как характер распределения при этом не поменялся совсем, что не повлияет на интерпретацию результатов в дальнейшем. 

**Также пропуски присутствуют и в столбце `floors_total`**, но их совсем немного, поэтому их также можно заменить медианным значением, вряд ли это сильно скажется на полученных результатах.

```python
data['floors_total'] = data['floors_total'].fillna(data['floors_total'].median())
data.hist('floors_total')
```

Распределение снова не изменилось, поэтому оставим заполненные пропуски таким образом. **Следующим шагом можно заполнить пропуски в столбце `living_area`:** например, можно посчитать медианное соотношение общей площади к жилой и умножить общую площадь на полученный коэффициент для заполнения пропусков. Данные будут очень приблизительные, но все же не случайные, так как основаны на общих тенденциях дата-сета. Подобное решение было бы неэффективным, если бы отсутствовало слишком много данных, но так как данные по общей площади в дата-сете представлены полностью, а по жилой отсутствует достаточно малая часть, то можно воспользоваться этой возможностью.

```python
living_area_coefficient = data['living_area'].median() / data['total_area'].median()
data['living_area'] = data['living_area'].fillna(data['total_area'] * living_area_coefficient)
data.hist('living_area')
```

Полученное распределенеи снова не сильно отличается от изначального, поэтому будем считать, что выбранный вариант заполнения пропусков оказался эффективным. **Далее перейдем к работе со столбцом `locality_name`:** к сожалению, заполнить пропуски тут вряд ли как-то получится в силу отсутствия каких-либо косвенных географических данных, а сами значения в столбце достаточно важны для проводимого анализа. В силу того, что отсутствующих данных не так много, можно удалить строки с отсутствием значений, так как эти данные не повлияют на проводимый анализ в силу небольшого количества, а если оставлять, то вряд ли окажутся полезными. Если бы отсутствующих данных оказалось много, то в таком случае целесообразно было бы обратиться к команде разработки: например, если это баг, то сообщить, или же сделать поле обязательным для заполнения, так как оно является достаточно значимым не только для анализа, но и для пользователей в целом.

```python
data = data.dropna(subset=['locality_name'])
```

**Со столбцом `kitchen_area`** предлагается не работать: во-первых, пропусков там сильно много, поэтому медианное значение может сказаться на интерпретируемых результатах, а косвенных данных, на основе которых можно построить закономерность - нет. Можно было бы пойти аналогичным способом, что и с жилой площадью, но проблема заключается в том, что не во всех квартирах есть кухня, поэтому результаты можно достаточно сильно исказить.

**Картографические данные** также предлагается не трогать в силу того, что пропуски в них выглядят достаточно закономерно: вполне вероятно, что данных по каким-либо объектам в системах просто нет, либо образовался баг при передаче данных из одного источника в другой. В любом случае пропусков достаточно много, поэтому на реальном проекте следовало бы обсудить этот вопрос, прежде всего, с командой разработки, а в рамках учебного проекта оставим данные в таком формате как есть, чтобы не испортить результаты исследования усредненными или медианными значениями.

**Интересными для работы являются пропуски в последнем столбце `days_exposition`:** пропусков слишком много и не похоже на то, чтобы эти данные не передавались в систему. Будем отталкиваться от того, что объявление еще активно, так как данный столбец показывает то, сколько дней объявление находилось в цикле продажи, но заполнять его при этом ничем не будем (как и удалять данные), чтобы избежать возможных искажений.

#### Изменение типов данных

Помимо пропусков в дата-сете также присутствуют некорректные типы данных, **в частности в столбцах: `first_day_exposition` и `is_apartment`**. Можно было бы также привести int и float типы к единому, но в рамках данного исследования это не требуется и не будет влиять на конечный результат анализа (например, если бы у нас были дискретные величины, то логичнее было бы представить их как int и так далее). **Единственный столбец, который может этого потребовать - `days_exposition`,** так как дни все-таки в большинстве случаев являются дискретными величинами, его и изменим.

**Столбец `first_day_exposition`** представлен в формате object, то есть является строкой, а должен быть датой, изменим тип данных при помощи соответствующего метода.

```python
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y.%m.%dT%H:%M:%S')
```

**Столбец `is_apartment`** тоже является объектом, но при этом должен быть булевым значением. 

```python
data['is_apartment'] = data['is_apartment'].astype('bool')
```

#### Обработка неявных дубликатов

**В столбце `locality_name`** помимо пропусков присутствуют также неявные дубликаты, которые могут сказаться не результатах проводимого анализа. Так как среди неявных дубликатов присутствуют одинаковые названия с е/ё, а также префиксами, то избавимся от того, и другого. Выведем список уникальных значений в столбце, чтобы удостовериться, что все неявные дубликаты были устранены.

```python
data['locality_name'] = data['locality_name'].replace('ё', 'е')
data['locality_name'] = data['locality_name'].replace( ['коттеджный ', 'станции ', 'садовое товарищество ', 'село ', 
                                                        'городской поселок ', 'городской посёлок ', 'деревня ', 'поселок ',
                                                        'посёлок ', 'городского типа ', 'при железнодорожной ', 'имени ',
                                                        'садоводческое некоммерческое товарищество '], '', regex=True)
data['locality_name'].unique()
```

- **во-первых, были охарактеризованы основные параметры объектов недвижимости:** они представлены 25-100 кв. метрами, при этом жилая площадь оказывается, как правило, с меньшей или большей жилой площадью. В основном квартирные дома представлены 4-8 этажами, а сами квартиры находятся на от 0 до 5-10 этажей. В основном объявления публикуются в будние дни, а также в послепраздничные месяцы (январь, май и декабрь самые холодные месяцы для продажи).
- **во-вторых, была проанализирована стоимость продажи квартир:** за основу были взяты медианные значения в силу присутствия аномальных показателей и смещенных средних оценок. Так, медианный срок продажи квартиры составляет 73 дни: 50% квартир продаются меньше этого срока, а другие 50% - больше.
- **в-третих, была проанализированы взаимосвязь цены квартиры с разными факторами.** Наибольшое влияние на цену оказывают всевозможные виды площади и количество комнат, а в меньшей - день и месяц.
- **в-четвертых, были изучены средние цены наиболее крупных с точки зрения публикации объявлений районов:** разбег составляет 60-100 тыс. руб., при этом самым доступным районом является Выборг, а дорогим - Санкт-Петербург.
- **в-пятых, была проанализирована взаимосвязь удаленности объекта недвижимости от центра города и цены за кв. метр:** зависимость оказалась обратной - значит, что при увеличении расстояния, то есть удаленности от центра, будет расти и цена. Тем не менее, существуют и прочие закономерности: вполне вероятно, что на цену этих квартир влияют и другие факторы, например, площадь, количество комнат и так далее.
